```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("filtro1.csv")
```

```python
#Filtro Y
filtro7=df[(df["PRECIO_UNITARIO"]<1000) & (df["FECHA_DOC"]=="2022-12-23")]
filtro7
```

```python
# filtro o 
filtro8=df[(df["PRECIO_UNITARIO"] < 1000)| (df["FECHA_DOC"]=="2022-12-23")]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("filtro8.csv")
```

```python
#filtro not
filtro9=df[~(df["PRECIO_UNITARIO"] > 1000) & ~(df["FECHA_DOC"]=="2022-12-23")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("filtro9.csv")
```
